Selenium Implementation 

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd

def scrape_page_selenium(url):
    # Configure the Selenium WebDriver
    options = Options()
    # options.add_argument("--headless")  # Run in headless mode
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    # options.add_argument('--disable-gpu')  # For Windows

    # Initialize the WebDriver
    driver = webdriver.Chrome(options=options)
    driver.get(url)

    # Wait for initial page load
    time.sleep(5)

    while True:
        # Scroll to the bottom of the page
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)

        # Wait for any dynamic content to load
        time.sleep(5)

        try:
            # Attempt to find and click the 'Load More' button
            load_more_button = driver.find_element(By.CSS_SELECTOR, ".ipc-see-more__text")
            load_more_button.click()

            # Wait for the new content to load
            time.sleep(5)
        except NoSuchElementException:
            # If the button is not found, break the loop
            break

    # Extract data
    items = driver.find_elements(By.CSS_SELECTOR, "ul.ipc-metadata-list--base > li")
    results = []
    for item in items:
        title_element = item.find_element(By.XPATH, ".//div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/a[1]")
        link_element = item.find_element(By.XPATH, ".//div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/a[1]")

        details = {
            'title': title_element.text if title_element else '',
            'link': link_element.get_attribute('href') if link_element else ''
        }
        results.append(details)

    driver.quit()
    return results

page_url = 'https://www.imdb.com/search/title/?title_type=tv_series&sort=release_date,desc&countries=TR'
page_data = scrape_page_selenium(page_url)
# Save the Dataframe to a csv file
df= pd.DataFrame(page_data)
df.to_csv('data.csv', index=False)


print(df)
```
